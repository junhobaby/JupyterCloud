# Files I/O
#A very common file format for python is text files. They can be easily manipulated using built-in-read()and write () functions. 

```python
f = open("foo.txt", "w") #writes to file
f.write("Hello!\n")
f.close()
```

```python
with open("foo.txt", "w") as f:  # writes to file
    f.write("Hello!\n")
# Closes automatically
```

```python
with open("foo.txt", "a") as f:  # amends existing file
    f.write("Another Hello!\n")
```

```python
with open("foo.txt", "r") as f: #reads from existing file 
    for line in f: 
        print (line.rstrip())     
```

# Python File Modes

#Mode	Description
#'r'	Open a file for reading. (default)
#'w'	Open a file for writing. Creates a new file if it does not exist or truncates the file if it exists.
#'x'	Open a file for exclusive creation. If the file already exists, the operation fails.
#'a'	Open for appending at the end of the file without truncating it. Creates a new file if it does not exist.
#'t'	Open in text mode. (default)
#'b'	Open in binary mode.
#'+'	Open a file for updating (reading and writing)

```python
with open ("foo.txt", "r") as f:
    for line in f:
        print(line.rstrip().split(' '))
```

# Pickling

Text files are convenient when data needs to be exchanged wiht other programs. However, if your work is limited to Python only, you might prefer to store Python data structures directly as binary files. This process is referred to as pickling

```python
import pickle as cPickle
```

```python
I = ["a", "list", "with", "text", [13,23,3.14], False]
```

```python
with open("my_list.pkl","wb") as f:
    cPickle.dump(I, f)
    
#dump: saving data into a pickle file (formats for storing data) usually in binary 01010101010
```

```python
with open("my_list.pkl", "rb") as f: 
    I = cPickle.load(f)

#load: when trying to open the content file what is inside in the jupyter file 
```

## Curve fitting

The data in the plot above looks like Gaussian. We can use the **curve_fit** function from Scipy to fit Gaussian to the data. The function can take any arbitrary function and fir it to the data. 

First, we need to define a general Gaussian function to fit. 

Gaussian = normal distribution in probability

```python
def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

#Now, fit its parameters to the data using **curve_fit** 
```
